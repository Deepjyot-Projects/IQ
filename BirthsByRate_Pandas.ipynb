```python
#Importing packages.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#reading csv into a dataframe
baby_names = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv')

#viewing dataframe
baby_names.head()
```

```python
#creating a smaller dataframe with only values we need
babies =  baby_names[['State', 'Count']]
```

```python
state_babycount = babies.groupby(['State'])["Count"].sum().reset_index()
```

```python
state_babycount.sort_values('Count',  ascending=False)
```

```python
print(state_babycount)
```

```python
# Plot the births by state
plt.figure(figsize=(12, 6))
plt.bar(state_babycount['State'], state_babycount['Count'])
plt.xlabel('State')
plt.ylabel('Total Births')
plt.title('Total Births by State')
plt.xticks(rotation=90)

# Annotate the top state on the plot
# plt.annotate(f"Top State: {top_state['State'].values[0]}", xy=(top_state['State'].values[0], top_state['Count'].values[0]),
#              xytext=(top_state['State'].values[0], top_state['Count'].values[0] + 500),
#              arrowprops=dict(arrowstyle='->'))

plt.show()
```
