```python
#Importing packages.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#reading csv into a dataframe
baby_names = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv')

#viewing dataframe
baby_names.head()
```

```python
#creating a smaller dataframe with only values we need
babies =  baby_names[['State', 'Count']]
```

```python
state_babyCount_df = babies.groupby(['State'])['Count'].sum().reset_index()


state_babyCount_df.sort_values('Count',  ascending=False)
```
